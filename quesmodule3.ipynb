```python
import quesmodule1
def print_bill(hotel_name, dish, number_of_plates):
    total_cost = quesmodule1.compute_bill(dish, number_of_plates)
    print("*******************************************************")
    print(hotel_name)
    print("********************************************************")
    print("Dish name: ", dish)
    print("Qty: ", number_of_plates)
    print("Total: ", total_cost)
    print("***********************************************************")
    discount = quesmodule1.compute_discount(total_cost)
    print("Discount: ", discount)
    amount_after_discount = total_cost - discount
    print("Amount after discount: ", amount_after_discount)
    print("***********************************************************")
    gst = quesmodule1.add_gst(amount_after_discount)
    print("GST: ", gst)
    print("**********************************************************")
    print("Payable Amount: ", amount_after_discount + gst)
    print("***********************************************************")
    print("HAVE A NICE DAY")
```
